```python
#coding=utf-8
import numpy as np  # 科学计算
import pandas as pd # 数据分析
from pandas import Series,DataFrame 
# 进行数据分析的可视化展示
import matplotlib.pyplot as plt
# 字体
from matplotlib.font_manager import FontProperties
font = FontProperties(fname='/Library/Fonts/Songti.ttc')

data_train = pd.read_csv("dataset/train.csv")
data_train # 训练数据
```

```python
data_train.info() # 数据的数据结构
data_train.describe() # 数据的描述性统计
```

```python

# 1、乘客的属性分布
fig = plt.figure()
fig.set_size_inches(12, 12)     #设置画布尺寸

# 获救人数
plt.subplot2grid((2,2),(0,0))   # 展示的图表在一个2x2的网格中，它的位置是左上，因为坐标是(0,0)
data_train.Survived.value_counts().plot(kind='bar')
plt.title("获救情况 (1为获救)", fontproperties=font)
plt.ylabel("人数",fontproperties=font)  

# 乘客等级人数分布
plt.subplot2grid((2,2),(0,1))  # 图表的位置在右上，因为坐标是(0,1)
data_train.Pclass.value_counts().plot(kind='bar')
plt.title("乘客等级人数分布情况",fontproperties=font)
plt.ylabel("人数",fontproperties=font)

# 乘客性别的人数分布
plt.subplot2grid((2,2),(1,0))
data_train.Sex[data_train.Survived == 1].value_counts().plot(kind='bar')
plt.title("乘客性别与获救人数的分布情况",fontproperties=font)
plt.ylabel("人数",fontproperties=font)

# 各等级的年龄分布
# plt.subplot2grid((2,2),(1,1))
# data_train.Age[data_train.Pclass == 1].plot(kind='kde')   
# data_train.Age[data_train.Pclass == 2].plot(kind='kde')
# data_train.Age[data_train.Pclass == 3].plot(kind='kde')
# plt.xlabel(u"年龄",fontproperties=font)# plots an axis lable
# plt.ylabel(u"密度",fontproperties=font) 
# plt.title(u"各等级的乘客年龄分布",fontproperties=font)
# plt.legend((u'First', u'Second',u'Three'),loc='best')

# 乘客等级与获救人数的情况
plt.subplot2grid((2,2),(1,1))
data_train.Pclass[data_train.Survived == 1].value_counts().plot(kind='bar')
# data_train.Survived[data_train.Pclass == 2].plot(kind='bar')
# data_train.Survived[data_train.Pclass == 3].plot(kind='bar')
# plt.xlabel(u"乘客等级",fontproperties=font)
plt.ylabel(u"人数",fontproperties=font)
plt.title(u"各等级的乘客获救人数",fontproperties=font)
# plt.legend((u'1', u'2',u'3'),loc='best')
plt.show()
```

```python
# 分析一下一等仓，年龄、性别对获救的影响
# 每个等级的获救比例
first_survived = data_train.Survived[data_train.Pclass == 1].value_counts()
print(first_survived)
second_survived = data_train.Survived[data_train.Pclass == 2].value_counts()
print(second_survived)
three_survived = data_train.Survived[data_train.Pclass == 3].value_counts()
print(three_survived)

# print("一等仓获救人数：", first_survived)
# 性别获救情况
# survived_male = data_train.Sex[data_train.Survived == 1].value_counts()
# print(survived_male)
# data_train.Sex[data_train.Survived == 1].value_counts().plot(kind='bar')
# plt.title("乘客等级人数分布情况",fontproperties=font)
# plt.ylabel("人数",fontproperties=font)
# plt.show()
```

```python
# 一等舱男性获救比例
first_male_survived = (data_train.Survived[data_train.Pclass == 1])[data_train.Sex == 'male'].value_counts()
print(first_male_survived)
print("一等舱男性获救比例", first_male_survived[1] / (first_male_survived[0] + first_male_survived[1]) ) 
# 一等舱女性性获救比例
first_female_survived = (data_train.Survived[data_train.Pclass == 1])[data_train.Sex == 'female'].value_counts()
print(first_female_survived)
print("一等舱女性获救比例", first_female_survived[1] / (first_female_survived[0] + first_female_survived[1]) ) 
```

```python
# 二等舱男性获救比例
second_male_survived = (data_train.Survived[data_train.Pclass == 2])[data_train.Sex == 'male'].value_counts()
print(second_male_survived)
print("二等舱男性获救比例", second_male_survived[1] / (second_male_survived[0] + second_male_survived[1]) ) 
# 二等舱女性性获救比例
second_female_survived = (data_train.Survived[data_train.Pclass == 2])[data_train.Sex == 'female'].value_counts()
print(second_female_survived)
print("二等舱女性获救比例", second_female_survived[1] / (second_female_survived[0] + second_female_survived[1]) ) 
```

```python
# 三等舱男性获救比例
three_male_survived = (data_train.Survived[data_train.Pclass == 3])[data_train.Sex == 'male'].value_counts()
print(three_male_survived)
print("三等舱男性获救比例", three_male_survived[1] / (three_male_survived[0] + three_male_survived[1]) ) 
# 三等舱女性性获救比例
three_female_survived = (data_train.Survived[data_train.Pclass == 3])[data_train.Sex == 'female'].value_counts()
print(three_female_survived)
print("三等舱女性获救比例", three_female_survived[1] / (three_female_survived[0] + three_female_survived[1]) ) 
```
